```python
def review_to_words(raw_review):
    #1.Remove HTML tags
    review_text = BeautifulSoup(raw_review).get_text()
    #2.Remove non-letters
    letters_only = re.sub("[^a-zA-Z]"," ",review_text)
    #3.Convert to lowercase than spliting
    words = letters_only.lower().split()
    #4.stopwords 
    stops = set(stopwords.words("english"))
    #5.Removing stopwords from words
    meaningful_words = [w for w in words if not w in stops]
    #6.Joining the sentence
    return(" ".join(meaningful_words))
    
```

```python
clean_review = review_to_words(train["review"][0])
print(clean_review)
```

```python
#Get the number of reviews based on dataframe column size
num_reviews = train["review"].size
#Initializing an empty list to store clean reviews
clean_train_reviews = []
#
for i in range(0,num_reviews):
    clean_train_reviews.append(review_to_words(train["review"][i]))
```

```python
print("Cleaning and parsing the training set movie reviews...\n")
clean_train_reviews = []
for i in range(0,num_reviews):
    if (i+1)%1000 == 0:
        print("Review %d of %d\n" % (i+1, num_reviews))
    clean_train_reviews.append(review_to_words(train["review"][i]))
```

```python
print("Creating the bag of words...\n")
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word",
                            tokenizer = None,
                            preprocessor = None,
                            stop_words = None,
                            max_features = 5000)

train_data_features = vectorizer.fit_transform(clean_train_reviews)

train_data_features = train_data_features.toarray()
```

```python
vocab = vectorizer.get_feature_names()
print(vocab)
```

```python
import numpy as np

dist = np.sum(train_data_features, axis = 1)

for tag, count in zip(vocab, dist):
    print(count, tag)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data_features, train["sentiment"], test_size=0.20)
```

```python
print("Training the random forest...")
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=100)

forest = forest.fit(X_train, y_train)
```

```python
preds = forest.predict(X_test)
```

```python
print(forest.score(X_test, y_test))
```

```python
from sklearn.metrics import f1_score, confusion_matrix
print("F1-score: {}".format(f1_score(preds, y_test)))
print("Confusion matrix:")
confusion_matrix(preds, y_test)
```
